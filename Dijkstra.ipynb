```python
# Graph Adjacency Matrix
largenum = 1000

Graph2 = [[largenum,2,5,largenum,largenum,largenum],
        [largenum,largenum,8,7,largenum,largenum],
        [largenum,largenum,largenum,2,4,largenum],
        [largenum,largenum,largenum,largenum,largenum,1],
        [largenum,largenum,largenum,6,largenum,3],
        [largenum,largenum,largenum,largenum,largenum,largenum]]
```

```python
class Dijkstra:
    
    def __init__(self,Graph):
        self.Graph = Graph
        self.V = len(Graph)
        
    def dijkstra(self):
        largenum = 1000
        V = self.V
        Graph = self.Graph
        
        # Queue of unvisited vertices
        Q = []
        for u in range(V):
            Q.append(u)
        
        # Initialize distance and previous lists
        dist = []
        prev = []
        for i in range(V):
            dist.append(largenum)
            prev.append(-1)
        dist[0] = 0
        
        # Loop while Q is not empty
        while Q:
            d_eval = []
            for i in Q:
                d_eval.append(dist[i])
            u = Q[d_eval.index(min(d_eval))]    # vertex in Q with min distance

            Q.remove(u)                         # remove the visited u

            neighbor = []                       # identify neighboing vertices of visited u
            for v in range(V):
                if Graph[u][v] != largenum:
                    neighbor.append(v)

            for v in neighbor:                  # for evry neighbor v: 
                newdist = dist[u] + Graph[u][v]

                if newdist<dist[v]:             # update distance and src if lower cost
                    dist[v] = newdist
                    prev[v] = u

        return dist,prev
    
    def generatepath(self):
        
        distances,previous = self.dijkstra()
        
        V = len(self.Graph)
        nodes = [i for i in range(V)]
        
        nxt = nodes[-1]
        path_id = [nxt]        
        while nxt > 0:    
            prev = previous[nxt]
            path_id.append(prev)
            nxt = prev
        
        # convert node index to node number
        path = []
        for i in range(len(path_id)):
            path.append(path_id[i]+1)

        return path      

                
```

```python
sol = Dijkstra(Graph1)
path = sol.generatepath()
print(path)
```
